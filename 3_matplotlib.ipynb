```python
#Where we use this subplots
#age.   experience.   bankbal     salary
#if i want to plot age and salary,  experience and salary,    bankbal and salary.    
a = np.random.randint(30,60,10)
exp = a+5
bb = np.random.randint(1000,10000,10)
sal = a*1000
plt.subplot(2,2,1)
plt.plot(a,sal,color='red')
plt.subplot(2,2,2)
plt.plot(exp,sal,color='blue')
plt.subplot(2,2,3)
plt.plot(bb,sal,color = 'green')
plt.subplot(2,2,4)
plt.plot(exp,bb,color = 'cyan')
plt.show()
```

```python
#dashed lines and stars
#let us change r-- and g*
a = np.random.randint(30,60,10)
exp = a+5
bb = np.random.randint(1000,10000,10)
sal = a*1000
plt.subplot(2,2,1)
plt.plot(a,sal,'r--')
plt.subplot(2,2,2)
plt.plot(exp,sal,color='blue')
plt.subplot(2,2,3)
plt.plot(bb,sal,'g*')
plt.subplot(2,2,4)
plt.plot(exp,bb,color = 'cyan')
plt.show()

```

```python
#bar graph
#bar graph represents data in the form of vertical or horizontal bars. It is useful to compare the quantities
#lets see employee id and salaries
import pandas as pd
df = pd.read_csv("empdata.csv")
```

```python
df
```

```python
#extracting empid and salary data into x and y axis
x = df['empid']
y = df['sal']
```

```python
#create bar graph
plt.bar(x,y,label = 'Employee data',color = 'red')
plt.xlabel('Employee id')
plt.ylabel('Employee salary')
plt.title('ABC India Pvt Ltd')
plt.legend()
plt.show()
```

```python
#from the above output, we can infer that employee id 1002 is drawing the highest salary and next to him is 1004
```

```python
#Histograms shows the distribution of values
#Histogram is similar to bar graph but it is useful to show values grouped in bins or intervals

```

```python
#Lets find the each employee age in a company and show inthe form of histogram to know how many employees are
# there in the range of 0-10,11-20....
emp_ages = [22,45,30,59,58,56,57,45,43,43,50,40,34,33,25,19]
bins = [0,10,20,30,40,50,60]
plt.hist(emp_ages,bins,label='ages',histtype='bar',rwidth=0.8,color='cyan')

#set labels
plt.xlabel('employee ages')
plt.ylabel('no. of employees')
plt.title('XYZ Group')
plt.legend()
plt.show()
```

```python
#Pie Chart
#It shows a circle that is divided into sectors and each sector represents portion of the whole
#lets display employees of different department in pie chart
slices = ['50','20','15','15']
depts = ['Sales','Production','HR','Finance']
cols = ['magenta','cyan','brown','gold']
plt.pie(slices,labels=depts,colors=cols, startangle = 90,explode=(0,0,0,0),shadow=True, autopct = '%.1f%%')
plt.title('WIPRO')
plt.legend()
plt.show()
```

```python
#To highlite a department
slices = ['50','20','15','15']
depts = ['Sales','Production','HR','Finance']
cols = ['magenta','cyan','brown','gold']
plt.pie(slices,labels=depts,colors=cols, startangle = 90,explode=(0,0.2,0,0),shadow=True, autopct = '%.1f%%')
plt.title('WIPRO')
plt.legend()
plt.show()
```
